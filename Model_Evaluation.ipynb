```python
import pandas as pd
import numpy as np
import ast
```

```python
predict = pd.read_excel('predictions.xlsx')
true_value = pd.read_excel('ground_truth.xlsx')
```

```python
predict
```

```python
def process(df, st):
    '''
    The function performs data processing
    df - dataframe
    st - name of column
    '''

    arr = []
    for i in range(50):
        if str(df[st][i]) == 'nan':
            arr.append(0)
        else:
            arr.append(np.array(ast.literal_eval(df[st][i])))
    
    return np.array(arr)
```

```python
predict_arr = process(predict, 'pred_bbox')
true_arr = process(true_value, 'truth_bbox')

```

```python
def IoU(boxAk, boxBk):
    
    """
    The function performs the calculation of the IoU indicator
    """
    boxA = []
    boxB = []
    
    boxA.append(abs(boxAk[0] - boxAk[2]/2))
    boxA.append(abs(boxAk[1] + boxAk[3]/2))
    boxA.append(abs(boxAk[0] + boxAk[2]/2))
    boxA.append(abs(boxAk[1] - boxAk[3]/2))
    
    boxB.append(abs(boxBk[0] - boxBk[2]/2))
    boxB.append(abs(boxBk[1] + boxBk[3]/2))
    boxB.append(abs(boxBk[0] + boxBk[2]/2))
    boxB.append(abs(boxBk[1] - boxBk[3]/2))
    
    
    left_max = max(boxA[0], boxB[0])
    top_max = max(boxA[1], boxB[1])
    right_min = min(boxA[2], boxB[2])
    bottom_min = min(boxA[3], boxB[3])
    

    inter = max(0, (right_min-left_max)) * max(0, top_max - bottom_min)
    Sa = (boxA[2] - boxA[0]) * (boxA[1] - boxA[3])
    Sb = (boxB[2] - boxB[0]) * (boxB[1] - boxB[3])

    
    union = Sa + Sb - inter
    res = inter / union
    
    return res
```

```python
#Найдем количество правильно детектированных голов кошек
#Примем IoU > 0.5 - правильно детектированная голова кошки
#Если на изображении нет кошки и модель не детектирует голову, результат правильный
#Расчет качества будет производиться по тремпоказателям: accuracy, precision, recall


iou_res = []
TP = 0 #True Positive
TN = 0 #True Negative
FP = 0 #False Positive
FN = 0 #False Negative

for i in range(50):
    if type(predict_arr[i]) == int and type(true_arr[i]) == int:
        iou_res.append(1)
        TN += 1
    elif type(predict_arr[i]) == int and type(true_arr[i]) != int:
        iou_res.append(0)
        FN += 1
    elif (type(predict_arr[i]) != int and type(true_arr[i]) == int):
        iou_res.append(0)
        FP += 1
    elif (type(predict_arr[i]) != int and type(true_arr[i]) != int) and (IoU(predict_arr[i], true_arr[i]) > 0.5):
        iou_res.append(1)
        TP += 1
    elif (type(predict_arr[i]) != int and type(true_arr[i]) != int) and (IoU(predict_arr[i], true_arr[i]) < 0.5):
        iou_res.append(1)
        FP += 1
    
        
```

```python
iou_res = np.array(iou_res)
print(iou_res)
print("True Positive: ", TP)
print("True Negative: ", TN)
print("False Positive: ", FP)
print("False Negative: ", FN)
```

```python
accuracy = sum(iou_res) / iou_res.shape[0]
precision = TP / (TP + FP)
recall = TP / (TP + FN)
```

```python
print("accuracy: ", accuracy)
print("precision: ", precision)
print("recall: ", recall)
```
